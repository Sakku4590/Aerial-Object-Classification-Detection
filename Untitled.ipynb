```python
from ultralytics import YOLO
```

```python
import os
os.environ['MLFLOW_TRACKING_URI'] = "file:///C:/Users/HP/runs/mlruns"

# Set the registry store to a local SQLite database file
# You must create the 'registry.db' file yourself.
os.environ['MLFLOW_REGISTRY_URI'] = "sqlite:///C:/Users/HP/runs/registry.db"

model = YOLO("yolov8s.pt")   # pre-trained model
results = model.train(
    data="object_detection_Dataset/data.yaml",
    epochs=10,
    imgsz=224,
    batch=16,
    patience=10,
    pretrained=True 
)
```

```python
model = YOLO("C:/Users/HP/runs/detect/train3/weights/best.pt")
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
names = model.names
```

```python
image_path = "D:/Aerial Object Classification & Detection/object_detection_Dataset/test/images/00ddd713b6f26a17_jpg.rf.0756de13413d353595f8a65e9300b032.jpg"
```

```python
results = model.predict(image_path, conf=0.25, verbose=False)

# Read original image
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
for r in results:
    for box in r.boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf[0])
        x1, y1, x2, y2 = map(int, box.xyxy[0])

        # Draw bounding box
        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)

        # Put label
        label = f"{names[cls_id]} {conf:.2f}"
        cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                    0.7, (255, 0, 0), 2)

        # Print clean output
        print(f"Class: {names[cls_id]}, Confidence: {conf:.2f}")
```

```python
%matplotlib inline
plt.figure(figsize=(8, 8))
plt.imshow(img)
plt.axis("off")
plt.show()
```
